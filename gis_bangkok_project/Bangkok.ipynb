```python
import pickle
import csv

list2 = []
f = open('table_bangkok.txt', 'rt')
try:
    reader = csv.reader(f)
    for row in reader:
        list2.append(row)
finally:
    f.close()
```

```python
list_to = []
for i in range(161):
    if i != 0:
        list_in=[]
        list_in.append(search('A_NAME_E: {}<BR>', list2[i][9])[0])
        list_in.append(search('T_NAME_E: {}<BR>', list2[i][9])[0])
        list_in.append(search('POPULATION: {}<BR>', list2[i][9])[0])
        list_in.append(search('HOUSE: {}<BR>', list2[i][9])[0])
        list_in.append(list2[i][10])
        list_to.append(list_in)
```

```python
with open('table_bkk.txt', 'a') as the_file:
    for i in list_to:
        the_file.write(str(i[0])+','+str(i[1])+','+str(i[2])+','+str(i[3])+','+str(i[4])+'\n')

the_file.close()
```
