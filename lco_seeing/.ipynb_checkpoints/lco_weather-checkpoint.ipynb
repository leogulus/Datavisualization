```python
from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
start_date = date(2017, 3, 1)
end_date = date(2017, 6, 23)

for single_date in daterange(start_date, end_date):
    day=single_date.strftime("%Y-%m-%d"); print day
    if day not in all_wind.keys():
        print 'date=', day
        print 'http://weather.lco.cl/clima/weather/Test/weathercalendar.html?date1=%s' % day 

        driver = webdriver.Chrome()
        driver.get('http://weather.lco.cl/clima/weather/Test/weathercalendar.html?date1=%s' % day)
        html1 = driver.page_source
        html2 = driver.execute_script("return document.documentElement.innerHTML;")
        driver.quit()

        soup = BeautifulSoup(html2, 'html.parser')
        print soup.find_all("path", class_='lineClay')
        try:
            path=soup.find_all("path", class_='lineClay')[0]
        except:
            print 'failed to download'
            continue
        wind=soup.find_all("path", class_='lineGusts')[0]

        try:
            seeing_pos=np.array([float(i.split(',')[1]) for i in path['d'][1:].split('L')])
        except: 
            seeing_pos=np.array([0,0])

        try:
            wind_pos=np.array([float(i.split(',')[1]) for i in wind['d'][1:].split('L')])
        except:
            wind_pos=np.array([0,0])
        all_seeing[day]=seeing_pos
        all_wind[day]=wind_pos
```

```python
fig,ax=plt.subplots(4,1, figsize=(18,14))

boxplot_all=[((2-0.6)/(-14.9917932367-95.0093226481)*all_seeing[i])+(1.8) for i in sorted(all_seeing.iterkeys())];

n=int(len(all_seeing.keys())/4)

a=n
b=2*n
c=3*n

boxplota=boxplot_all[:a]
ax[0].boxplot(boxplota);
ax[0].axhline(y=1,ls='--')
ax[0].set_xticks(np.arange(1,len(boxplota)+1,1))
ax[0].set_xticklabels([i[5:] for i in sorted(all_seeing.iterkeys())[:a]])
ax[0].set_ylim(0,2)
ax[0].set_ylabel('Seeing (arcsecond)')

boxplotb=boxplot_all[a:b]
ax[1].boxplot(boxplotb);
ax[1].axhline(y=1,ls='--')
ax[1].set_xticks(np.arange(1,len(boxplotb)+1,1))
ax[1].set_xticklabels([i[5:] for i in sorted(all_seeing.iterkeys())[a:b]])
ax[1].set_ylim(0,2)
ax[1].set_ylabel('Seeing (arcsecond)')

boxplotc=boxplot_all[b:c]
ax[2].boxplot(boxplotc);
ax[2].axhline(y=1,ls='--')
ax[2].set_xticks(np.arange(1,len(boxplotc)+1,1))
ax[2].set_xticklabels([i[5:] for i in sorted(all_seeing.iterkeys())[b:c]])
ax[2].set_ylim(0,2)
ax[2].set_ylabel('Seeing (arcsecond)')

boxplotd=boxplot_all[c:]
ax[3].boxplot(boxplotd);
ax[3].axhline(y=1,ls='--')
ax[3].set_xticks(np.arange(1,len(boxplotd)+1,1))
ax[3].set_xticklabels([i[5:] for i in sorted(all_seeing.iterkeys())[c:]])
ax[3].set_ylim(0,2)
ax[3].set_ylabel('Seeing (arcsecond)')

ax[0].set_title('Seeing at Megellan Telescope')

# plt.savefig('test.png')
plt.savefig('test2.pdf',dpi=360)
```
