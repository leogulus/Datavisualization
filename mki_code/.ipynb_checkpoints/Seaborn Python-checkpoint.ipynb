```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
def make_plot(**kwargs):
    return plt.plot([1,2,3],[2,3,4],'o--',**kwargs)   
```

```python
make_plot()
```

```python
with sns.plotting_context('talk'):
    fig=plt.figure()
    make_plot()
```

```python
sns.set(context='talk',style='ticks',font='serif',palette='muted')
make_plot()
```
